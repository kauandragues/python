## Exercícios com funções

Crie uma função que multiplica todos os argumentos não nomeados recebidos e retorne o total para uma variável mostrando o valor da variável.

```python
def produto(*args):
    total = 1
    for numero in args:
        total *= numero

    return total

print(produto(1,2,3,4,5))
```

Crie uma função que diz se o número é impar ou par.

```python
def eh_impar_ou_par(num):
    mutiplo_de_dois = num % 2 == 0

    if mutiplo_de_dois:
        return f'{num} é par'
    
    return f'{num} é impar'
    
print(eh_impar_ou_par(4))
print(eh_impar_ou_par(7))
```
