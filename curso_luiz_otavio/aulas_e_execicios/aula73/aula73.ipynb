Higher Order Functions

Funções de primeira classe

Funções são tratadas como qualquer outra variável ou valor. Isso significa que funções podem ser atribuídas a variáveis, passadas como argumentos para outras funções e retornadas por funções

Academicamente, os termos Higher Order Functions e First-Class Functions têm significados diferentes.

    Higher Order Functions - Funções que podem receber e/ou retornar outras funções

    First-Class Functions - Funções que são tratadas como outros tipos de dados comuns (strings, inteiros, etc...)

```python
def saudacao(mensagem):
    return mensagem

v1 = saudacao
v2 = v1

print(v2(saudacao('V2 Bom dia')))
```

```python
def ola(mensagem, nome):
    return f'{mensagem}, {nome}'

def executa(funcao, *args):
    return funcao(*args)

v = ola
argumentos = 'Olá', 'Kauã'
print(executa(v,'Bom dia', 'Rodrigo'))
print(executa(v,*argumentos))
```
