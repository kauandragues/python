# Introdução às Generator functions

```python
generator = (n for n in range(100))
print(next(generator))
print(next(generator))
print(next(generator))
```

```python
def generator(indice=0):
    yield 1
    print('Continuando..')
    yield 2
    print('Continuando..')
    yield 3
    print('Continuando..')
    yield 4
    print('Continuando..')
    yield 5
    return 'Acabou'

gen = generator(0)
print(next(gen))
print('OUTRO CÓDIGO')
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
print(next(gen))
```

```python
def generator(indice=0):
    yield indice
    indice+=1
    yield indice
    indice+=1
    yield indice
    indice+=1
    yield indice
    indice+=1
    yield indice
    return 'Acabou'


gen = generator(0)
for i in gen:
    print(i)
```

```python
def generator(indice=0, maximo=100):
    while True:
        indice+=1
        yield indice

        if indice >= maximo:
            return 'Acabou'


gen = generator(0, 10)
for i in gen:
    print(i)
```
