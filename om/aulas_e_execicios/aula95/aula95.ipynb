```python
def divide(n, d):
    try:
        return n/d
    except ZeroDivisionError:
        print("fazer algum log para saber sobre esse erro no futuro")
        raise
    

print(divide(8,0))
```

```python
def nao_aceita_nao_numero(numero):
    if isinstance(numero, (float, int)):
        return True
    raise ("Você tentou dividir um não número")

def nao_aceita_zero(d):
    if d == 0:
        raise ZeroDivisionError
    
    return True

def divide(n, d):
    nao_aceita_zero(d)
    nao_aceita_nao_numero(n)
    nao_aceita_nao_numero(d)
    return n/d
    

print(divide(8,0))
```
