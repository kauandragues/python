## `isinstance()`
Serve para saber se um objeto é de determinado tipo.

```python
lista = [
    'a', 1, 1.1, True, [0,1,2], (1,2), {0,1}, {'nome':'Kauã'}
]

for item in lista:
    print(item, isinstance(item, set))

    if isinstance(item, (int, float)):
        print(item, 'de novo')
```
