o * e *args (empacotamento e desempacotamento)

## Empacotamento

```python
x, y, *resto = 1, 2, 3, 4
print(x, y, resto)
```

```python
def soma (x, y):
    return x + y

soma = soma(1,2,4,5,6,7)
print(soma)
```

```python
def soma(*args):
    soma = 0
    for i in args:
        soma += i
    
    return soma

soma = soma(1,2,3,4,5,6,7)
print(soma)
```

```python
print(sum((1,2,3,4,5,6,7))) # 28
print(sum(1,2,3,4,5,6,7)) # soma um iteravel, e aqui ele entende que todos esses números são argumentos da função
```

```python
def soma(*args):
    return sum(args)

soma = soma(1,2,3,4,5,6,7)
print(soma)
```

```python
numeros = 1,2,3,4,5,6,6,7 # isso é a mesma coisa que (1,2,3,4,5,6,6,7), é apenas outra maneira de fazer uma tupla
print(numeros)
print(*numeros)
soma = sum(*numeros) # desempacotou como vários argumentos da função
print(soma)
```
