## Exercícios

Crie funções que duplicam, triplicam e quadruplicam o número recebido como parâmetro.

```python
def duplicar(num:float) -> float:
    return num*2

def triplicar(num:float) -> float:
    return num*3

def quadruplicar(num:float) -> float:
    return num*4

print(f"duplicar 4: {duplicar(4.0):.2f}")
print(f"triplicar 3: {triplicar(3.0):.2f}")
print(f"quadruplicar 11: {quadruplicar(11.0):.2f}")
```

```python
def criar_funcoes_multiplica(multiplicador):
    def multiplicar(numero):
        return multiplicador*numero
    
    return multiplicar

duplicar = criar_funcoes_multiplica(2)
triplicar = criar_funcoes_multiplica(3)
quadruplicar = criar_funcoes_multiplica(4)

print(f"duplicar 4: {duplicar(4)}")
print(f"triplicar 3: {triplicar(3)}")
print(f"quadruplicar 11: {quadruplicar(11)}")
```
