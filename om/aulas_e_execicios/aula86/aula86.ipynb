## Dictionary Comprehension e Set Comprehension

```python
produto = {
    'nome':'Caneta Azul',
    'preco':2.5,
    'categoria': 'Escritório',
}

for chave, valor in produto.items():
    print(chave,valor)
```

```python
produto = {
    'nome':'Caneta Azul',
    'preco':2.5,
    'categoria': 'Escritório',
}

novo_produto = {chave:valor.upper() if isinstance(valor, str) else valor for chave, valor in produto.items() if chave == 'categoria'}
print(novo_produto)
```

```python
lista = [
    ('a', 'valor a'),
    ('b', 'valor b'),
    ('c', 'valor c'),
    ('d', 'valor d'),
]

dicionario = {chave:valor for chave, valor in lista}
print(dicionario)
```

Para fazer um set comprehension, basta não passar chave.

```python
s1 = {i for i in range(10)}
print(s1)
```
