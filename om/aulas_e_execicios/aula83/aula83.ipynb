## Empacotamento e desempacotamento de dicionários

```python
a, b = 1, 2
print(a, b)
a, b = b, a
print(a, b)
```

```python
pessoa = {
    'nome': 'Aline',
    'sobrenome':'Souza',
}

a, b = pessoa.values()

(a1, a2), (b1, b2) = pessoa.items()
print(a1, a2)
print(b1, b2)
print(a, b)
```

```python
pessoa = {
    'nome': 'Aline',
    'sobrenome':'Souza',
}

for chave, valor in pessoa.items():
    print(chave, valor)
```

```python
pessoa = {
    'nome': 'Aline',
    'sobrenome':'Souza',
}

dados_pessoa = {
    'idade': 24,
    'altura': 1.6,
}
pessoa_completa = {**pessoa, **dados_pessoa}
print(pessoa_completa)
```

```python
def mostro_argumentos_nomeados(*args, **kwargs):
    print(args)
    print(kwargs)

mostro_argumentos_nomeados(1,2,4,nome='Kauã', idade=20)
```

```python
pessoa = {
    'nome': 'Aline',
    'sobrenome':'Souza',
}

def mostrar_argumentos_nomeados(**kwargs):
    print(kwargs)

mostrar_argumentos_nomeados(**pessoa)
```
