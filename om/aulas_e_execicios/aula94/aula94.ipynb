## Try, except, else e finally
# https://docs.python.org/pt-br/3/library/exceptions.html#built-in-exceptions

```python
try:
    print('Tenta fazer')
except NameError:
    print('Caso aconteça um erro')
else:
    print('Caso não dê erro')
finally:
    print('Acontece não importa o que aconteça')
```
