## Try, except, else e finally

```python
try:
    a = 18
    b = 2
    c = a / b

    d = [1,2,3,4]
    e = d[10]
except ZeroDivisionError:
    print('Dividiu por zero')
except NameError:
    print('Nome b não está definido')
except (TypeError, IndexError) as error:
    print(error)
    print(error.__class__.__name__)
except Exception:
    print('Erro deconhecido')
```
