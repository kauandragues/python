## Generator expression, Iterables e Iterators

Iterator não sabe quase nada sobre o objeto iterável, ele apenas conhece quem será o próximo elemento. Ele te ajuda a navegar um objeto com os métodos `iter()` e `next()`.

```python
iterable = ['Eu', 'Tenho', '__iter__']
iterator = iterable.__iter__() ## tem __iter__ e __next__
print(next(iter(iterable)))
print(next(iterator))
print(next(iterator))
print(next(iterator))
```

Generator são função que sabem faltar em um determinado ponto. Todo Generator é um Iterator. É um iterável acessável apenas pelo iterator dele, ele não armazena na memória todos os valores do objeto ao mesmo tempo, o que o torna ótimo para atividades que não precisam de dados armazenados constantemente na memória, e só aqueles utilizados pelo iterator.

Por conta disso:
- Não é possível acessar o tamanho de um generator, já que ele apenas sabe qual é o valor que o iterator está apontando.
- Não é possível acessar itens fora da ordem do iterator, é um por vez, seguindo a ordem de iteração.

```python
import sys

lista = [n for n in range(100)]
generator = (n for n in range(10))
print(sys.getsizeof(lista))
print(sys.getsizeof(generator))
```
