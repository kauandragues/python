## Dicionários em Python

Dicionários são estruturas de dados do tipo par de "chave" e "valor".

Chaves podem ser consideradas como o "índice" que vimos na lista e podem ser de tipos imutáveis como: `str`, `int`, `float`, `bool`, `tuple`, etc.

O valor pode ser de qualquer tipo, incluindo outro dicionário.

Usamos as chaves `{ }` ou a classe `dict` para criar dicionários.

Imutáveis: `str`, `int`, `float`, `bool`, `tuple`

Mutáveis: `dict`, `list`

```python
pessoa = dict(nome='Kauã', sobrenome='de Andrade')

pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'de Andrade',
    'idade': 18,
    'altura': 1.75,
    'endereços':[
        {'rua':'exemplo de rua', 'número':123},
        {'rua': 'exemplode rua2', 'número':321},    
    ]
}

# print(pessoa['nome'])
# print(pessoa['sobrenome'])

for chave in pessoa:
    print(pessoa[chave])
print(pessoa)
```

Métodos úteis dos dicionários em Python

- `len()` - quantas chaves
- `keys()` - iterável com chaves
- `values()` - iterável com os valores
- `items()` - iterável com chaves e valores
- `setdefault()` - adiciona valor se a chave não existe
- `copy()` - retorna uma cópia rasa (shallow copy)
- `get()` - obtém uma chave
- `pop()` - apaga um item com a chave especificada (del)
- `popitem()` - apaga o último item adicionado
- `update()` - atualiza um dicionário com outro

```python
pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'chave substituida depois',
    'sobrenome':'Andrade',
}

print(pessoa.__len__())
print(len(pessoa))
```

```python
pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'Andrade',
}

print(list(pessoa.keys())) # retorna todas as chaves

lista = []
for chave in pessoa:
    lista.append(chave)

print(lista)
```

```python
pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'Andrade',
}

print(list(pessoa.values())) # retorna todas as chaves

lista = []
for chave in pessoa:
    lista.append(pessoa[chave])

print(lista)
```

```python
pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'Andrade',
}

print(list(pessoa.items())) # retorna todas as chaves

lista = []
for chave in pessoa:
    lista.append((chave, pessoa[chave]))

print(lista)
```

```python
pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'Andrade',
}
pessoa.setdefault('idade',0)
print(pessoa['idade'])
```

```python
import copy

pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'Andrade',
    'l1': [0, 1, 2],
}
pessoa_copia_rasa = pessoa.copy()
# aponta para o mesmo lugar na memória, não faz uma cópia rasa
# para fazer uma cópia rasa é preciso usar o copy(), ele faz uma cópia rasa
# ou seja, quaiquer tipos de dados que forem mutáveis e estiverem dentro do dict, irão continuar referenciando o mesmo lugar na memória
# para fazer uma deep copy é precisa importar copy e usar o método copy.copy(dicionário_que_deseja_copiar)
pessoa_copia_rasa['nome'] = 'Pedro'
pessoa_copia_rasa['l1'][0] = 1000

pessoa_copia_profunda = copy.deepcopy(pessoa)
pessoa_copia_profunda['nome'] = 'Maria'
pessoa_copia_profunda['l1'][0] = 3042

print('original:',pessoa)
print('copia rasa:', pessoa_copia_rasa)
print('copia profunda:', pessoa_copia_profunda)
```

```python
pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'Andrade',
}
print(pessoa.get('nome','mensagem pra caso não exista a chave'))
```

```python
pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'Andrade',
}

nome = pessoa.pop('nome')
print(nome)
print(pessoa)
```

```python
pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'Andrade',
}

nome = pessoa.popitem()
print(nome)
print(pessoa)
```

```python
pessoa = {
    'nome': 'Kauã',
    'sobrenome': 'Andrade',
}

pessoa.update({
    'nome':'João',
})

pessoa.update(sobrenome='Miranda')

tupla = ('idade', 19),

lista = ['peso', 70],

pessoa.update(tupla)
pessoa.update(lista)

print(pessoa)
```
