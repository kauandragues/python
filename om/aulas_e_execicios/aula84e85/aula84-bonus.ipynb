## Como funciona a list comprehension?

```python
numeros = [1,2,3,4,5,6]
novos_numeros = numeros
novos_numeros[0] = 15 
# a lista numeros e novos_numeros apontam para o mesmo local na memória, não são variáveis independentes
# se uma delas for alterar a outra também recebe a alteração
print(numeros)
print(novos_numeros)
```

Se quisermos copiar uma lista sem fazer com que elas referenciem umas as outras, temos que fazer uma shallow copy ou deep copy.
- Shallow Copy copia os valores de uma variável para outra variável, mas não cria cópias únicas de valores que são mutáveis e identados.
- Já o Deep Copy faz uma copia completamente única dos valores de uma variável para outra, criam-se variáveis realmente independentes.

Podemos fazer um loop de atribuir cada um dos elementos para uma nova variável

```python
numeros = [1,2,3,4,5,6]
novos_numeros = []
for numero in numeros:
    novos_numeros.append(numero)
novos_numeros[0] = 15 

print(numeros)
print(novos_numeros)
```

Mas isso apenas funciona para listas que não possuem valores identados ou mutáveis:

```python
numeros = [[1,2,3],[4,5],[6,7]]
novos_numeros = []
for numero in numeros:
    novos_numeros.append(numero)

novos_numeros[0][1] = 34

print(numeros) # as lista numero e novos_numeros são indenpendentes, porém é uma shallow copy
print(novos_numeros) # as listas dentro delas não são indenpendetes e referenciam umas as outras
```

```python
numeros = [[1,2,3],[4,5],[6,7]]
novos_numeros = [x for x in numeros] # esse list comprehension faz o mesmo que o for anterior
novos_numeros[0][1] = 34
print(novos_numeros)
print(numeros)
```

```python
numeros = [[1,2,3],[4,5],[6,7]]
novos_numeros = [[y for y in x] for x in numeros] # esse list comprehension consegue criar uma deepcopy
novos_numeros[0][1] = 34
print(novos_numeros)
print(numeros)
```
