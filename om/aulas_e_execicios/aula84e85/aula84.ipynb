## List comprehension e com mais de 1 for
As list comprehensions são uma forma rápida para criar listas a partir de iteráveis.

```python
lista = [numero*2 for numero in range(10)]
print(lista)
```

```python
quantidade_entradas = int(input())
entrada = [input() for _ in range(quantidade_entradas)]

print(quantidade_entradas, entrada)
```

```python
print(list(range(10)))
```

```python
lista = []
for numero in range(10):
    lista.append(numero)

print(lista)
```

## Mapeamento de dados em list comprehension

```python
produtos = [
    {'nome':'p1', 'preco':20,},
    {'nome':'p2', 'preco':10,},
    {'nome':'p3', 'preco':30,},
]


novos_produtos = [{**produto, 'preco':produto['preco']*1.5} if produto['preco'] == 10 else {**produto} for produto in produtos]
print(novos_produtos) 
print(produtos)
```

```python
lista = [num for num in range(10) if num < 5]
print(lista)
```

```python
lista = list()
for n in range(3):
    for num in range(3):
        lista.append((n, num))

print(lista)
```

```python
lista = list()
lista = [(x,y) for x in range(3) for y in range(3)]
print(lista)
```

```python
lista = list()
lista = [[(x,y) for y in range(3)] for x in range(3)]
print(lista)
```
